```python
df.drop(["fruit_subtype"],axis=1, inplace=True)
```

```python
df
```

```python
X = df[["mass","width","height","color_score"]]
y = df['fruit_label']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
```

```python
from inspect import classify_class_attrs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("Accuracy : ", acc)
print( classification_report(y_test, y_pred))
```

```python
acc = knn.score(X_test, y_test)

print("The accuracy of our model is: %.2f" % round(acc,2))
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with different colors for each class
scatter = ax.scatter(X_train["mass"], X_train["width"], X_train["height"], c=y_train, marker='o', s=100)

# Set labels for each axis
ax.set_xlabel('Mass')
ax.set_ylabel('Width')
ax.set_zlabel('Height')

# Add a legend manually
classes = set(y_train)
legend_labels = [f'Class {cls}' for cls in classes]
ax.legend(legend_labels, title="Classes")

plt.show()
```
